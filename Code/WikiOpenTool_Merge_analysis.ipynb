```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import requests
import json
```

```python
Group_S = ["Glasgow School of Art","Royal Conservatoire of Scotland","Scotland's Rural College"]
Group_M = ["Abertay University","Queen Margaret University","Glasgow Caledonian University","Edinburgh Napier University",
      "University of the West of Scotland","Robert Gordon University","University of the Highlands and Islands"]
Group_A = ["University of Edinburgh","University of Glasgow","University of St Andrews","University of Aberdeen"]
Group_C = ["University of Dundee","Heriot-Watt University","University of Strathclyde","University of Stirling"]
```

```python
all_S_data = []

for uni in Group_S :
    
    data = pd.read_csv(f"../Data/alt_oa_n_data/n_{uni}_AltOA.csv")
    print(uni,data.shape[0])
    columns = data.columns.tolist()
    new_order = columns[-2:] + columns[:-2]
    data = data[new_order]
    all_S_data.append(data)

combined_S_data = pd.concat(all_S_data)

print(combined_S_data.shape[0])
combined_S_data.head()
```

```python
all_M_data = []

for uni in Group_M :
    
    data = pd.read_csv(f"../Data/alt_oa_n_data/n_{uni}_AltOA.csv")
    print(uni,data.shape[0])
    columns = data.columns.tolist()
    new_order = columns[-2:] + columns[:-2]
    data = data[new_order]
    all_M_data.append(data)

combined_M_data = pd.concat(all_M_data)

print(combined_M_data.shape[0])
combined_M_data.head()
```

```python
all_A_data = []

for uni in Group_A :
    
    data = pd.read_csv(f"../Data/alt_oa_n_data/n_{uni}_AltOA.csv")
    print(uni,data.shape[0])
    columns = data.columns.tolist()
    new_order = columns[-2:] + columns[:-2]
    data = data[new_order]
    all_A_data.append(data)

combined_A_data = pd.concat(all_A_data)

print(combined_A_data.shape[0])
combined_A_data.head()
```

```python
all_C_data = []

for uni in Group_C :
    
    data = pd.read_csv(f"../Data/alt_oa_n_data/n_{uni}_AltOA.csv")
    print(uni,data.shape[0])
    columns = data.columns.tolist()
    new_order = columns[-2:] + columns[:-2]
    data = data[new_order]
    all_C_data.append(data)

combined_C_data = pd.concat(all_C_data)


print(combined_C_data.shape[0])
combined_C_data.head()
```

```python
# Merge all the dataframes
merged_df = pd.concat([combined_A_data,combined_C_data,combined_M_data,combined_S_data], ignore_index=True)

# Reset the index (or remove it entirely)
merged_df.reset_index(drop=True, inplace=True)

# Save the merged result to a new CSV file (optional)
# merged_df.to_csv(f"../Data/merged_all_output.csv",index=False)
# print("Done download!")
merged_df.tail()
```
